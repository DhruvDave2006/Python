### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter a year"))
if year % 400 == 0:
    print(f"{year} is leap year")
elif year % 4 == 0 and year % 100 != 0:
    print(f"{year} is leap year")
else:
    print(f"{year} is not leap year")
```

### 06) WAP in python to display the name of the day according to the number given by the user.

```python
choice = int(input("Enter input form 1 to 7 : "))
match choice:
    case 1:
        print("Sunday")
    case 2:
        print("Monday")
    case 3:
        print("Tuesday")
    case 4:
        print("Wednesday")
    case 5:
        print("Thursday")
    case 6:
        print("Friday")
    case 7:
        print("Saturday")
    case _:
        print("Invalid input")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))
choice = int(input("Enter choice from 1 to 4 : "))

match choice:
    case 1:
        print(f"Add is : {num1 + num2}")
    case 2:
        print(f"Sub is : {num1 - num2}")
    case 3:
        print(f"Mul is : {num1 * num2}")
    case 4:
        print(f"Div is : {num1 // num2}")
    case _:
        print("Invalid Input")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
sub1, sub2, sub3, sub4, sub5 = map(int, input("Enter marks of 5 subjects").split())
per = (sub1 + sub2 + sub3 + sub4 + sub5) / 5

if per > 70:
    print("Distinction")
elif per > 60:
    print("First Class")
elif per > 45:
    print("Second class")
elif per > 35:
    print("Pass Class")
else:
    print("Fail")
```
