### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input())
add = 0
for i in range(1,n+1):
    if i % 2 != 0:
        add += i
    else:
        add -= i
print("Sum Of Series :",add)
```

### 13) WAP to print sum of digits of given number.

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n= int(input())
originalNum = n
count=0;
while n!=0:
    rem=n%10 
    n//=10
    count+=1
    
n = originalNum
sum = 0
while n!=0:
    rem=n%10
    sum+=rem**count
    n//=10

if sum==originalNum:
    print("Is Armstrong number")
else:
    print("Is not Armstrong number")
```
