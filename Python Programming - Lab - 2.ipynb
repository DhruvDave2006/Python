### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
c = float(input())
f = float(input())

feh = (c * (9 / 5) + 32)
deg = (f - 32) * (5 / 9)

print("Fahrenheit =", feh)
print("Degree =", deg)
```
